```python
# 배급 수수료 비율의 기본값을 두고 최종 정산 매출(만원)을 계산해 봅니다.
def total_with_fee(p, fee_rate=0.1):  # fee_rate를 안 주면 0.1(=10%) 사용
    total = p - (p * fee_rate)
    return total  # 계산한 값을 돌려줍니다.

revenue = 5500
print("기본 수수료 적용 후 매출:", total_with_fee(revenue))
print("수수료 20% 적용 후 매출:", total_with_fee(revenue, 0.2))
```

```python
# 여러 편의 영화 매출에서 평균(평균 매출, 만원)을 구해 값을 '반환'합니다.
revenues = [5200, 4800, 6100]  # 아주 작은 예시 리스트

def average_revenue(rlist):
    # 합을 구하고 개수로 나눕니다.
    total = 0
    # for는 이전에 배운 반복문을 그대로 씁니다.
    for r in rlist:
        total = total + r
    avg = total / len(rlist)
    return avg  # 계산 결과를 함수 바깥으로 돌려줍니다.

avg_revenue = average_revenue(revenues)
print("평균 매출(만원):", avg_revenue)
```

```python
# 함수: 흥행 기준 매출(int)과 영화 매출 정보(dict)를 함께 받아 처리합니다.
def check_hit(limit_revenue, movie_data):
	print("=== 흥행 여부 확인 ===")
	print("기준 매출:", limit_revenue, "만원\n")
	# 딕셔너리의 key는 영화 제목, value는 매출입니다.
	for title in movie_data:
		revenue = movie_data[title]
		print("영화:", title)
		print("매출:", revenue, "만원")
		# 기준 이상이면 흥행, 미만이면 평범
		if revenue >= limit_revenue:
			print("→ 흥행 성공")
		else:
			print("→ 아직 평범")
		print()  # 한 줄 띄우기

# 예시 실행
hit_limit = 5000  # 흥행 기준 매출(만원)
movies = {
	"서울의 밤": 5200,
	"블루 나이트": 3100,
	"라스트 퀘스트": 4950
}

check_hit(hit_limit, movies)
```

```python
# 흥행 성공 영화만 따로 모아 반환하는 함수
def get_hit_list(limit_revenue, movie_data):
	result = []  # 비어 있는 리스트 생성
	for title in movie_data:
		revenue = movie_data[title]
		if revenue >= limit_revenue:
			result.append(title)  # 조건 만족 시 제목 저장
	return result

hit_list = get_hit_list(5000, movies)
print("흥행 성공 영화 목록:", hit_list)
```

```python
# 목표: CSV 파일을 "문자(텍스트)"로 읽고, 줄 단위로 나누어 일부만 출력해 본다.
# 배경: 영화 박스오피스(영화 정보) CSV가 같은 폴더에 있다고 가정한다.
# 파일명: KC_KOBIS_BOX_OFFIC_MOVIE_INFO_202412.csv  (인코딩: utf-8)

# 1) 파일 열기: open(파일경로, "r"=읽기모드, encoding=문자코드)
f = open("KC_KOBIS_BOX_OFFIC_MOVIE_INFO_202412.csv", "r", encoding="utf-8")

# 2) 전체 내용을 하나의 긴 문자열로 읽기
text = f.read()

# 3) 파일 닫기 (컴퓨터 자원 반환)
f.close()

# 4) 줄 단위로 나누기: 줄바꿈 기준으로 쪼개서 리스트(여러 칸) 만들기
lines = text.split("\n")

# 5) 맨 앞 3줄만 미리보기 (헤더 포함)
print("총 줄 수:", len(lines))           # 몇 줄인지 출력
print("=== 미리보기 3줄 ===")
print(lines[0])                          # 1줄 (헤더, 예: NO,MOVIE_NM,...)
print(lines[1])                          # 2줄 (1번 영화 정보)
print(lines[2])                          # 3줄 (2번 영화 정보)
```
