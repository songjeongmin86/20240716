```python
import pandas as pd
import glob
import os

path = "/content/sample_data/"
# Filter files to only include those relevant to punctuality data
files = [f for f in glob.glob(os.path.join(path, "*.csv")) if "Data_raw_punctuality" in f]

use_cols = [
    "DATDEP",
    "PLANNED_TIME_ARR",
    "PLANNED_TIME_DEP",
    "DELAY_ARR",
    "DELAY_DEP",
    "PTCAR_LG_NM_NL"
]

df_list = []
for f in files:
    temp = pd.read_csv(f, usecols=use_cols)
    df_list.append(temp)

df = pd.concat(df_list, ignore_index=True)
print(df.head())
print(df.info())

df["DATDEP"] = pd.to_datetime(df["DATDEP"], format="%d%b%Y")
df["DATE"] = df["DATDEP"].dt.strftime("%Y-%m-%d")
df["YEAR"] = df["DATDEP"].dt.year
df["MONTH"] = df["DATDEP"].dt.month
df["DAY"] = df["DATDEP"].dt.day
print(df.head())

df["DELAY_ARR_MIN"] = df["DELAY_ARR"] / 60
df["DELAY_DEP_MIN"] = df["DELAY_DEP"] / 60

df["IS_LATE_ARR"] = (df["DELAY_ARR"] > 0).astype(int)
df["IS_LATE_DEP"] = (df["DELAY_DEP"] > 0).astype(int)

by_month = df.groupby("MONTH").agg(
    avg_delay_arr_min=("DELAY_ARR_MIN", "mean"),
    avg_delay_dep_min=("DELAY_DEP_MIN", "mean"),
    late_rate_arr=("IS_LATE_ARR", "mean"),
    late_rate_dep=("IS_LATE_DEP", "mean")
).reset_index()

print(by_month)

df[["STATION", "ZONE"]] = df["PTCAR_LG_NM_NL"].str.split("-", n=1, expand=True)

by_station = df.groupby("STATION").agg(
    avg_delay_arr_min=("DELAY_ARR_MIN", "mean"),
    avg_delay_dep_min=("DELAY_DEP_MIN", "mean"),
    late_rate_arr=("IS_LATE_ARR", "mean"),
    late_rate_dep=("IS_LATE_DEP", "mean")
).reset_index()

print(by_station)

top5_station = by_station.sort_values(by="avg_delay_dep_min", ascending=True
).head(5)

print(top5_station)
```
