```python
# 영화 박스오피스 데이터를 간단한 '객체(오브젝트)'로 다루기 위한 클래스 예제
# CSV에 있을 법한 항목을 작게 흉내냅니다: 제목, 매출, 상영관 수, 러닝타임

class Movie:
	# 클래스는 '설계도'입니다. 실제 영화 한 편(객체)을 만들기 위한 틀입니다.
	def __init__(self, title, revenue, screens, running_time):
		# self는 "지금 만들어지는 그 객체 자신"을 가리킵니다.
		self.title = title              # 영화 제목(문자열)
		self.revenue = revenue          # 매출(만원 단위, 정수 또는 실수)
		self.screens = screens          # 상영관 수(정수)
		self.running_time = running_time  # 러닝타임(분, 정수)

	def summary(self):
		# 객체에 저장된 정보를 보기 좋게 한 줄로 출력합니다.
		print(self.title + " / " + str(self.revenue) + "만원 / "
		      + str(self.screens) + "관 / " + str(self.running_time) + "분")

# 설계도(클래스)로부터 실제 객체 2개를 만들어 봅니다.
movie1 = Movie("서울의 밤", 52000, 120, 128)
movie2 = Movie("블루 나이트", 46500, 85, 110)

# 각 객체가 가진 정보를 출력해 봅니다.
movie1.summary()
movie2.summary()
```

```python
# '함수'는 클래스 밖에 있는 동사 같은 코드 조각입니다.
def is_over_target(revenue, target):
	# 매출이 기준 이상인지 True/False로 알려줍니다.
	return revenue >= target

# '메소드'는 클래스 안에 있는 함수입니다.
# 아래는 메소드를 하나 더 추가한 예입니다.
class MovieWithCheck:
	def __init__(self, title, revenue):
		self.title = title
		self.revenue = revenue

	def is_over(self, target):
		# 내 매출(self.revenue)이 기준 이상인지 판단합니다.
		return self.revenue >= target

movie3 = MovieWithCheck("별빛의 노래", 4550)  # 단위: 만원
print(is_over_target(movie3.revenue, 5000))  # 함수 사용
print(movie3.is_over(5000))                  # 메소드 사용
```

```python
print(movie3.is_over(4000))  # True가 나오면 정상
```

```python
print(is_over_target(4700, 4800))  # False
```

```python
movie4 = MovieWithCheck("바다의 기억", 6100)
print(movie4.is_over(6000))  # True
print(movie4.is_over(7000))  # False
```

```python
# 변수(variable)는 하나의 값을 저장합니다.
# 예: 숫자, 문자열 등 간단한 정보
revenue = 5200
title = "서울의 밤"

print("제목:", title)
print("매출(만원):", revenue)

# 속성(attribute)은 객체 안에 저장된 변수입니다.
# 객체마다 자기만의 속성을 가집니다.

class Movie:
	def __init__(self, title, revenue, screens):
		self.title = title      # 객체의 제목 속성
		self.revenue = revenue  # 객체의 매출 속성
		self.screens = screens  # 객체의 상영관 수 속성

# 객체(movie1)를 하나 만들어 봅니다.
movie1 = Movie("블루 나이트", 4800, 90)

# 객체 안의 속성에 접근하려면 '객체이름.속성'을 사용합니다.
print("제목:", movie1.title)
print("매출(만원):", movie1.revenue)
print("상영관 수:", movie1.screens)

# 속성값을 바꿀 수도 있습니다.
movie1.revenue = 5000
print("변경된 매출:", movie1.revenue)
```

```python
# 상속(inheritance)은 기존 클래스를 '물려받아' 새 클래스를 만드는 것입니다.
# 부모 클래스(기존 클래스)의 속성과 메소드를 그대로 쓸 수 있습니다.

class Movie:
	def __init__(self, title, revenue, rating):
		self.title = title
		self.revenue = revenue
		self.rating = rating
	def summary(self):
		print("제목:", self.title, "/ 매출(만원):", self.revenue, "/ 평점:", self.rating)

# 새 클래스: Movie를 상속받은 PremiumMovie
class PremiumMovie(Movie):
	# 추가 기능만 정의하면 됩니다.
	def show_premium(self):
		print(self.title, "은(는) 프리미엄 상영관에서 상영 중입니다.")

# Movie 기능을 그대로 사용할 수 있습니다.
movie1 = PremiumMovie("서울의 밤", 9500, 9.1)
movie1.summary()       # 부모 클래스의 메소드 사용
movie1.show_premium()  # 자식 클래스에 새로 추가된 메소드
```

```python
# Override(오버라이드)는 부모 클래스의 메소드를 '같은 이름으로 다시 정의'하는 것입니다.
# 이렇게 하면 새 클래스의 방식으로 바꿔서 작동합니다.

class MovieWithDate(Movie):
	def __init__(self, title, revenue, rating, release_date):
		# 부모의 __init__을 그대로 쓰고 싶으면 super()를 사용합니다.
		super().__init__(title, revenue, rating)
		self.release_date = release_date
	# 부모의 summary()를 다시 정의(Override)
	def summary(self):
		print("제목:", self.title,
		      "/ 매출(만원):", self.revenue,
		      "/ 평점:", self.rating,
		      "/ 개봉일:", self.release_date)

# 사용 예시
movie2 = MovieWithDate("한강의 노을", 8700, 8.7, "2025-03-01")
movie2.summary()  # 개봉일까지 함께 출력
```
