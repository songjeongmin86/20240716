```python
# 목표: CSV 파일을 "문자(텍스트)"로 읽고, 줄 단위로 나누어 일부만 출력해 본다.
# 배경: 영화 박스오피스(영화 정보) CSV가 같은 폴더에 있다고 가정한다.
# 파일명: KC_KOBIS_BOX_OFFIC_MOVIE_INFO_202412.csv  (인코딩: utf-8)

# 1) 파일 열기: open(파일경로, "r"=읽기모드, encoding=문자코드)
f = open("KC_KOBIS_BOX_OFFIC_MOVIE_INFO_202412.csv", "r", encoding="utf-8")

# 2) 전체 내용을 하나의 긴 문자열로 읽기
text = f.read()

# 3) 파일 닫기 (컴퓨터 자원 반환)
f.close()

# 4) 줄 단위로 나누기: 줄바꿈 기준으로 쪼개서 리스트(여러 칸) 만들기
lines = text.split("\n")

# 5) 맨 앞 3줄만 미리보기 (헤더 포함)
print("총 줄 수:", len(lines))           # 몇 줄인지 출력
print("=== 미리보기 3줄 ===")
print(lines[0])                          # 1줄 (헤더, 예: NO,MOVIE_NM,...)
print(lines[1])                          # 2줄 (1번 영화 정보)
print(lines[2])                          # 3줄 (2번 영화 정보)
```

```python
# 폴더 안의 여러 박스오피스 CSV 파일을 하나로 합치기
import os

folder = "boxoffice"
files = os.listdir(folder)

merged = ""   # 합칠 내용을 담을 문자열
first = True  # 첫 번째 파일만 헤더 포함

for f in files:
    if f.endswith(".csv"):
        path = folder + "/" + f
        text = open(path, "r", encoding="utf-8").read()
        lines = text.split("\n")
        if first:
            merged = text           # 첫 파일은 전부
            first = False
        else:
            merged = merged + "\n" + "\n".join(lines[1:])  # 헤더 제외 추가

open("merged_boxoffice.csv", "w", encoding="utf-8").write(merged)
print("merged_boxoffice.csv 저장 완료")
```
