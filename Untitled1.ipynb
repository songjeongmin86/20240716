```python
import pandas as pd
import sqlite3
import os
import sys
import tkinter as tk
from tkinter import filedialog, messagebox
from openpyxl.styles import Font
import shutil
import subprocess

#ë°˜ë“œì‹œ python 3.10 ë²„ì „ìœ¼ë¡œ ëŒë ¤ì£¼ì„¸ìš”. 3.13ë²„ì „ì¼ ê²½ìš° ëŒì•„ê°€ì§€ ì•ŠìŠµë‹ˆë‹¤.

# --- 1ë‹¨ê³„ í•¨ìˆ˜ ---

def select_excel_file():
    print("ğŸ“‚ ì—‘ì…€ íŒŒì¼ ì„ íƒì„ ìœ„í•œ íŒì—…ì°½ì„ ì—½ë‹ˆë‹¤...")
    root = tk.Tk(); root.withdraw()
    return filedialog.askopenfilename(
        title="ì²˜ë¦¬í•  ì—‘ì…€ íŒŒì¼ì„ ì„ íƒí•˜ì„¸ìš”",
        filetypes=[("Excel files", "*.xlsx *.xls"), ("All files", "*.*")]
    )


def add_customers_from_excel_to_db(excel_path, db_path):
    print("âœ… 1ë‹¨ê³„: ì—‘ì…€ ë°ì´í„°ë¥¼ ë°ì´í„°ë² ì´ìŠ¤ì— ì¶”ê°€í•˜ëŠ” ì‘ì—…ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    conn = None
    try:
        conn = sqlite3.connect(db_path)
        cur = conn.cursor()
        print(f"ğŸ”— ë°ì´í„°ë² ì´ìŠ¤ '{os.path.basename(db_path)}'ì— ì—°ê²°ë˜ì—ˆìŠµë‹ˆë‹¤.")
        cur.execute('''
            CREATE TABLE IF NOT EXISTS customers (
                receipt_number TEXT PRIMARY KEY, customer_name TEXT, phone_number TEXT
            )
        ''')
        print("ğŸ“œ 'customers' í…Œì´ë¸”ì„ í™•ì¸í•˜ê³ , ì—†ìœ¼ë©´ ìƒì„±í•˜ê² ìŠµë‹ˆë‹¤.")

        df = pd.read_excel(excel_path, usecols=['ê³ ê°ëª…', 'íœ´ëŒ€í°', 'ì ‘ìˆ˜ë²ˆí˜¸'], dtype=str)
        print(f"ğŸ“„ '{os.path.basename(excel_path)}' íŒŒì¼ì—ì„œ {len(df)}ê°œì˜ ë°ì´í„°ë¥¼ ì½ì—ˆìŠµë‹ˆë‹¤.")

        original_count = len(df)
        df.dropna(subset=['ì ‘ìˆ˜ë²ˆí˜¸'], inplace=True)
        df = df[df['ì ‘ìˆ˜ë²ˆí˜¸'].astype(str).str.strip() != '']
        dropped_count = original_count - len(df)
        if dropped_count > 0:
            print(f"ğŸ§¹ 'ì ‘ìˆ˜ë²ˆí˜¸'ê°€ ë¹„ì–´ìˆëŠ” {dropped_count}ê°œì˜ í–‰ì„ DB ì¶”ê°€ ëŒ€ìƒì—ì„œ ì œì™¸í–ˆìŠµë‹ˆë‹¤.")

        new_data_count = 0
        for _, row in df.iterrows():
            # DBì— ì¶”ê°€ (ì´ë¯¸ ìˆëŠ” ì ‘ìˆ˜ë²ˆí˜¸ëŠ” ë¬´ì‹œë¨)
            cur.execute('INSERT OR IGNORE INTO customers VALUES (?, ?, ?)', (row['ì ‘ìˆ˜ë²ˆí˜¸'], row['ê³ ê°ëª…'], row['íœ´ëŒ€í°']))
            if cur.rowcount > 0:
                new_data_count += 1
        conn.commit()
        print(f"ğŸ’¾ ë°ì´í„°ë² ì´ìŠ¤ì— ì´ {new_data_count}ê°œì˜ ì‹ ê·œ ê³ ê° ì •ë³´ê°€ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return True
    except Exception as e:
        print(f"âŒ 1ë‹¨ê³„ ì˜¤ë¥˜: {e}"); root = tk.Tk(); root.withdraw(); messagebox.showerror("1ë‹¨ê³„ ì˜¤ë¥˜", f"ì—‘ì…€ ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜:\n\n{e}")
        return False
    finally:
        if conn: conn.close()
        print("ğŸ”— ë°ì´í„°ë² ì´ìŠ¤ ì—°ê²°ì„ ì¢…ë£Œí–ˆìŠµë‹ˆë‹¤.")
        print("âœ… 1ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

# --- 2ë‹¨ê³„ ê´€ë ¨ í•¨ìˆ˜ ---
def process_and_classify_excel(excel_path, db_path):
    print(f"âœ… 2ë‹¨ê³„: '{os.path.basename(excel_path)}' íŒŒì¼ ë¶„ë¥˜ ì‘ì—…ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    output_filename = None
    try:
        df = pd.read_excel(excel_path, dtype=str).fillna('')
        phone_col = 'íœ´ëŒ€í°' if 'íœ´ëŒ€í°' in df.columns else 'ì „í™”ë²ˆí˜¸'
        if phone_col not in df.columns: raise KeyError("'íœ´ëŒ€í°' ë˜ëŠ” 'ì „í™”ë²ˆí˜¸' ì—´ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

        conn = sqlite3.connect(db_path)
        cur = conn.cursor()
        print(f"ğŸ” DBì— ì—°ê²°í•˜ì—¬ ê° í–‰ì˜ ê³¼ê±° ì´ë ¥ì„ í™•ì¸í•©ë‹ˆë‹¤.")

        df['ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] = ''
        re_customer_count = 0
        unanswered_count = 0

        for index, row in df.iterrows():
            cur.execute('SELECT 1 FROM customers WHERE customer_name = ? AND phone_number = ? AND receipt_number != ? LIMIT 1',
                        (row['ê³ ê°ëª…'], row[phone_col], row['ì ‘ìˆ˜ë²ˆí˜¸']))
            is_re_complaint = cur.fetchone() is not None

            if not row['ë‹µë³€']:
                df.at[index, 'ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] = 'ë¯¸ë‹µë³€'; unanswered_count += 1
            if is_re_complaint:
                df.at[index, 'ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] = 'ì¬ë¯¼ì›'; re_customer_count += 1
                if not row['ë‹µë³€']: unanswered_count -= 1

        conn.close()
        print(f"ğŸ“Š ë¶„ë¥˜ ì™„ë£Œ: ì¬ë¯¼ì› {re_customer_count}ê±´, ë¯¸ë‹µë³€ {unanswered_count}ê±´")

        base, ext = os.path.splitext(excel_path)
        output_filename = f"{base}_2ë‹¨ê³„ì™„ë£Œ{ext}"
        df.to_excel(output_filename, index=False)
        print(f"ğŸ’¾ ê²°ê³¼ê°€ '{os.path.basename(output_filename)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return output_filename
    except Exception as e:
        print(f"âŒ 2ë‹¨ê³„ ì˜¤ë¥˜: {e}"); root = tk.Tk(); root.withdraw(); messagebox.showerror("2ë‹¨ê³„ ì˜¤ë¥˜", f"ë¶„ë¥˜ ì‘ì—… ì¤‘ ì˜¤ë¥˜:\n\n{e}")
        return None
    finally:
        print("âœ… 2ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

# --- 3ë‹¨ê³„ í•¨ìˆ˜ ---
def apply_final_formatting(input_path, db_path):
    print(f"âœ… 3ë‹¨ê³„: '{os.path.basename(input_path)}' íŒŒì¼ì— ìµœì¢… ì„œì‹ ì ìš©ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    try:
        df = pd.read_excel(input_path, dtype=str).fillna('')

        if 'ë“±ë¡ì¼' in df.columns:
            df['ë“±ë¡ì¼'] = pd.to_datetime(df['ë“±ë¡ì¼'], errors='coerce')
            print("ğŸ“… 'ë“±ë¡ì¼' ì—´ì„ ë‚ ì§œ í˜•ì‹ìœ¼ë¡œ ë³€í™˜í–ˆìŠµë‹ˆë‹¤.")
        else:
            print("âš ï¸ 'ë“±ë¡ì¼' ì—´ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ì²« ë²ˆì§¸ ë¯¼ì› ì œì™¸ ë¡œì§ì„ ê±´ë„ˆëœë‹ˆë‹¤.")

        re_complaint_mask = (df['ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] == 'ì¬ë¯¼ì›')

        if 'ë“±ë¡ì¼' in df.columns and not df['ë“±ë¡ì¼'].isna().all():
            df['ê³ ê°í‚¤'] = df['ê³ ê°ëª…'] + '_' + df['íœ´ëŒ€í°']
            # í˜„ì¬ íŒŒì¼ ê¸°ì¤€, ê³ ê°ë³„ ì²« ë¯¼ì›ì˜ ì¸ë±ìŠ¤ë¥¼ ì°¾ìŒ
            first_complaint_indices = df.groupby('ê³ ê°í‚¤')['ë“±ë¡ì¼'].idxmin()

            indices_to_exclude = []
            conn = None
            try:
                print("ğŸ” ë°ì´í„°ë² ì´ìŠ¤ë¥¼ í™•ì¸í•˜ì—¬ 'ì§„ì§œ' ì²« ë¯¼ì› ì—¬ë¶€ë¥¼ ê²€ì¦í•©ë‹ˆë‹¤...")
                conn = sqlite3.connect(db_path)
                cur = conn.cursor()

                # í˜„ì¬ íŒŒì¼ì˜ ì²« ë¯¼ì› í›„ë³´ë“¤ì— ëŒ€í•´ DB ì´ë ¥ì„ í•˜ë‚˜ì”© í™•ì¸
                first_complaint_df = df.loc[first_complaint_indices]
                for idx, row in first_complaint_df.iterrows():
                    cur.execute('SELECT COUNT(*) FROM customers WHERE customer_name = ? AND phone_number = ?', (row['ê³ ê°ëª…'], row['íœ´ëŒ€í°']))
                    db_count = cur.fetchone()[0]

                    # DB ì´ë ¥ì´ 2ê±´ ë¯¸ë§Œì¸ ê³ ê°ë§Œ 'ì²« ë¯¼ì› ì œì™¸' ê·œì¹™ì„ ì ìš©
                    if db_count < 2:
                        indices_to_exclude.append(idx)

                print(f"ğŸ¯ DB ì´ë ¥ ê²€ì¦ ì™„ë£Œ: 'ì²« ë¯¼ì›' í›„ë³´ {len(first_complaint_indices)}ê±´ ì¤‘ {len(indices_to_exclude)}ê±´ì„ ì¬ë¯¼ì›ì—ì„œ ìµœì¢… ì œì™¸í•©ë‹ˆë‹¤.")

            except Exception as e:
                print(f"âš ï¸ DB ê²€ì¦ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}. ê¸°ì¡´ ë°©ì‹(íŒŒì¼ ê¸°ì¤€)ìœ¼ë¡œ ì²« ë¯¼ì› ì œì™¸ ë¡œì§ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.")
                indices_to_exclude = list(first_complaint_indices)
            finally:
                if conn: conn.close()

            re_complaint_mask = re_complaint_mask & ~df.index.isin(indices_to_exclude)

        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜1'] = '(ë¬¼ì )ê¸°íƒ€'
        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜2'] = 'ê¸°íƒ€'
        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜3'] = 'ê¸°íƒ€'
        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜4'] = 'ì¬ë¯¼ì›'
        print("âœï¸ 'ì¬ë¯¼ì›' í•­ëª©ì— ëŒ€í•œ í…ìŠ¤íŠ¸ ì±„ìš°ê¸°ë¥¼ ì™„ë£Œí–ˆìŠµë‹ˆë‹¤.")

        base, ext = os.path.splitext(input_path)
        base = base.replace('_2ë‹¨ê³„ì™„ë£Œ', '')
        output_filename = f"{base}_3ë‹¨ê³„ì™„ë£Œ{ext}"

        with pd.ExcelWriter(output_filename, engine='openpyxl') as writer:
            df.drop(columns=['ê³ ê°í‚¤'], inplace=True, errors='ignore')
            df.to_excel(writer, index=False, sheet_name='Sheet1')
            worksheet = writer.sheets['Sheet1']
            red_font = Font(color="FF0000")
            target_cols = ['ë‚´ìš©ë¶„ë¥˜1', 'ë‚´ìš©ë¶„ë¥˜2', 'ë‚´ìš©ë¶„ë¥˜3', 'ë‚´ìš©ë¶„ë¥˜4']
            col_indices = {col: list(df.columns).index(col) + 1 for col in target_cols if col in df.columns}

            for row_idx in df.index[re_complaint_mask]:
                excel_row = row_idx + 2
                for col_num in col_indices.values():
                    worksheet.cell(row=excel_row, column=col_num).font = red_font

        print("ğŸ¨ 'ì¬ë¯¼ì›' í•­ëª©ì— ëŒ€í•œ ê¸€ììƒ‰ ì„œì‹ ì ìš©ì„ ì™„ë£Œí–ˆìŠµë‹ˆë‹¤.")
        full_path = os.path.abspath(output_filename)
        print(f"ğŸ’¾ 3ë‹¨ê³„ ê²°ê³¼ê°€ '{os.path.basename(output_filename)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return full_path
    except Exception as e:
        print(f"âŒ 3ë‹¨ê³„ ì˜¤ë¥˜: {e}"); root = tk.Tk(); root.withdraw(); messagebox.showerror("3ë‹¨ê³„ ì˜¤ë¥˜", f"ì„œì‹ ì ìš© ì¤‘ ì˜¤ë¥˜:\n\n{e}")
        return None
    finally:
        print("âœ… 3ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

def step_4_run_predictor_and_process(step3_output_path):
    print(f"âœ… 4ë‹¨ê³„: ì™¸ë¶€ ì˜ˆì¸¡ ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰ ë° ê²°ê³¼ ë³‘í•©ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    final_output_path = None
    predictor_input_path = None
    predictor_output_path = None

    try:
        script_dir = os.path.dirname(os.path.abspath(__file__))
        predictor_script_name = 'predictor_for_categorize.py'
        script_to_run_path = os.path.join(script_dir, 'module', predictor_script_name)

        predictor_input_filename = 'VOC_ëª©ë¡.xlsx'
        predictor_output_filename = 'predicted_data.xlsx'
        predictor_input_path = os.path.join(script_dir, predictor_input_filename)
        predictor_output_path = os.path.join(script_dir, predictor_output_filename)

        shutil.copy(step3_output_path, predictor_input_path)
        print(f"ğŸ“‹ '{os.path.basename(step3_output_path)}'ë¥¼ '{predictor_input_filename}'ìœ¼ë¡œ ë³µì‚¬í–ˆìŠµë‹ˆë‹¤.")

        if not os.path.exists(script_to_run_path):
            raise FileNotFoundError(f"ì‹¤í–‰í•  ìŠ¤í¬ë¦½íŠ¸ '{script_to_run_path}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤!")

        print(f"ğŸš€ '{predictor_script_name}' ìŠ¤í¬ë¦½íŠ¸ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤...")
        env = os.environ.copy()
        env.update({'TF_CPP_MIN_LOG_LEVEL': '2', 'TF_ENABLE_ONEDNN_OPTS': '0'})

        process = subprocess.Popen(
            [sys.executable, "-u", script_to_run_path],
            stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
            cwd=script_dir, env=env
        )

        for line_bytes in process.stdout:
            try:
                line = line_bytes.decode('utf-8')
            except UnicodeDecodeError:
                line = line_bytes.decode('cp949', errors='replace') # í˜¹ì‹œ ëª¨ë¥¼ ì˜¤ë¥˜ ë°©ì§€
            print(line, end='', flush=True)

        process.wait()
        if process.returncode != 0:
            raise subprocess.CalledProcessError(process.returncode, [sys.executable, script_to_run_path])

        print(f"\nğŸ‰ '{predictor_script_name}' ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰ ì™„ë£Œ!")

        if not os.path.exists(predictor_output_path):
            raise FileNotFoundError(f"ì˜ˆì¸¡ ìŠ¤í¬ë¦½íŠ¸ê°€ ê²°ê³¼ íŒŒì¼('{predictor_output_filename}')ì„ ìƒì„±í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤!")

        print(f"ğŸ“Š ì›ë³¸ íŒŒì¼ê³¼ ì˜ˆì¸¡ ê²°ê³¼ íŒŒì¼ì„ ì½ìŠµë‹ˆë‹¤.")
        df_original = pd.read_excel(step3_output_path, dtype=str).fillna('')
        df_predicted = pd.read_excel(predictor_output_path, dtype=str).fillna('')

        print("âœï¸ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ì›ë³¸ ë°ì´í„°ì— ë³‘í•©í•˜ê³  ê°€ê³µí•©ë‹ˆë‹¤...")
        df_original['ë‚´ìš©ë¶„ë¥˜'] = df_predicted.get('ë‚´ìš©ë¶„ë¥˜', '')
        df_original['ë‚´ìš©ë¶„ë¥˜(2ì°¨)'] = df_predicted.get('ë‚´ìš©ë¶„ë¥˜(2ì°¨)', '')

        re_complaint_mask = (df_original['ë‚´ìš©ë¶„ë¥˜4'] == 'ì¬ë¯¼ì›')

        for index, row in df_original.iterrows():
            if re_complaint_mask[index]: continue

            primary_cat, secondary_cat = str(row.get('ë‚´ìš©ë¶„ë¥˜', '')), str(row.get('ë‚´ìš©ë¶„ë¥˜(2ì°¨)', ''))
            source_string = None

            if primary_cat and primary_cat != 'ê¸°íƒ€': source_string = primary_cat
            elif primary_cat == 'ê¸°íƒ€' and secondary_cat: source_string = secondary_cat

            if source_string:
                parts = [p.strip() for p in source_string.split(' > ')]
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜1'] = parts[0] if len(parts) > 0 else ''
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜2'] = parts[1] if len(parts) > 1 else ''
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜3'] = parts[2] if len(parts) > 2 else ''
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜4'] = parts[3] if len(parts) > 3 else ''

        print("ë³‘í•© ë° ê°€ê³µ ì™„ë£Œ.")

        base, ext = os.path.splitext(step3_output_path)
        final_output_path = base.replace('_3ë‹¨ê³„ì™„ë£Œ', '_4ë‹¨ê³„ì™„ë£Œ') + ext
        df_original.drop(columns=['ë‚´ìš©ë¶„ë¥˜', 'ë‚´ìš©ë¶„ë¥˜(2ì°¨)'], inplace=True, errors='ignore')
        df_original.to_excel(final_output_path, index=False)

        print(f"ğŸ’¾ 4ë‹¨ê³„ ê²°ê³¼ê°€ '{os.path.basename(final_output_path)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return final_output_path

    except (FileNotFoundError, subprocess.CalledProcessError, Exception) as e:
        error_type = type(e).__name__
        print(f"âŒ 4ë‹¨ê³„ ì˜¤ë¥˜ ({error_type}): {e}")
        root = tk.Tk(); root.withdraw(); messagebox.showerror(f"4ë‹¨ê³„ {error_type}", f"ì˜¤ë¥˜ ë°œìƒ:\n\n{e}")
        return None
    finally:
        if predictor_input_path and os.path.exists(predictor_input_path):
            os.remove(predictor_input_path); print(f"ğŸ§¹ ì„ì‹œ ì…ë ¥ íŒŒì¼ ì‚­ì œ ì™„ë£Œ.")
        if predictor_output_path and os.path.exists(predictor_output_path):
            os.remove(predictor_output_path); print(f"ğŸ§¹ ì„ì‹œ ê²°ê³¼ íŒŒì¼ ì‚­ì œ ì™„ë£Œ.")
        print("âœ… 4ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")


def step_5_run_secondary_processing(step4_output_path):
    print(f"âœ… 5ë‹¨ê³„: 2ì°¨ ë°ì´í„° ê°€ê³µ ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    final_output_path = None

    try:
        script_dir = os.path.dirname(os.path.abspath(__file__))
        script_name = 'second_process.py'
        script_to_run_path = os.path.join(script_dir, 'module', script_name)

        if not os.path.exists(script_to_run_path):
            raise FileNotFoundError(f"ì‹¤í–‰í•  ìŠ¤í¬ë¦½íŠ¸ '{script_to_run_path}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤!")

        base, ext = os.path.splitext(step4_output_path)
        final_output_path = base.replace('_4ë‹¨ê³„ì™„ë£Œ', '_5ë‹¨ê³„ì™„ë£Œ_ìµœì¢…') + ext

        print(f"ğŸš€ '{script_name}' ìŠ¤í¬ë¦½íŠ¸ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤...")

        process = subprocess.Popen(
            [sys.executable, "-u", script_to_run_path, step4_output_path, final_output_path],
            stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
            cwd=script_dir
        )

        for line_bytes in process.stdout:
            try:
                line = line_bytes.decode('utf-8')
            except UnicodeDecodeError:
                line = line_bytes.decode('cp949', errors='replace')
            print(line, end='', flush=True)

        process.wait()
        if process.returncode != 0:
            raise subprocess.CalledProcessError(process.returncode, [sys.executable, script_to_run_path])

        if not os.path.exists(final_output_path):
            raise FileNotFoundError("2ì°¨ ê°€ê³µ ìŠ¤í¬ë¦½íŠ¸ê°€ ê²°ê³¼ íŒŒì¼ì„ ìƒì„±í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤!")

        full_path = os.path.abspath(final_output_path)
        success_msg = f"ëª¨ë“  ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!\n\nìµœì¢… ê²°ê³¼ê°€ '{os.path.basename(final_output_path)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.\n\nì €ì¥ ìœ„ì¹˜:\n{full_path}"
        print(f"\nğŸ’¾ {success_msg}")
        root = tk.Tk(); root.withdraw(); messagebox.showinfo("ìµœì¢… ì™„ë£Œ", success_msg)
        return final_output_path

    except (FileNotFoundError, subprocess.CalledProcessError, Exception) as e:
        error_type = type(e).__name__
        print(f"âŒ 5ë‹¨ê³„ ì˜¤ë¥˜ ({error_type}): {e}")
        root = tk.Tk(); root.withdraw(); messagebox.showerror(f"5ë‹¨ê³„ {error_type}", f"ì˜¤ë¥˜ ë°œìƒ:\n\n{e}")
        return None
    finally:
        print("âœ… 5ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")


def confirm_step(step_number, description):
    print(f"ğŸ¤” {step_number}ë‹¨ê³„ ì§„í–‰ ì—¬ë¶€ë¥¼ ë¬»ëŠ” íŒì—…ì°½ì„ ì—½ë‹ˆë‹¤...")
    root = tk.Tk(); root.withdraw()
    root.attributes('-topmost', True)
    return messagebox.askyesno(
        title=f"{step_number}ë‹¨ê³„ ì§„í–‰ í™•ì¸",
        message=f"{step_number-1}ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n\nì´ì–´ì„œ {step_number}ë‹¨ê³„({description})ë¥¼ ì§„í–‰í•˜ì‹œê² ìŠµë‹ˆê¹Œ?"
    )

# --- ë©”ì¸  ---
if __name__ == "__main__":
    try:
        script_dir = os.path.dirname(os.path.abspath(__file__))
    except NameError:
        script_dir = os.getcwd()
    db_file_name = 'ëˆ„ì _ê³ ê°_2025.db'
    db_file_path = os.path.join(script_dir, db_file_name)

    print("--- ğŸ“‚ ë°ì´í„°ë² ì´ìŠ¤ íŒŒì¼ í™•ì¸ ---")
    if not os.path.exists(db_file_path):
        print(f"â„¹ï¸ '{db_file_name}' íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ìƒˆ íŒŒì¼ì„ ìƒì„±í•©ë‹ˆë‹¤.")
    else:
        print(f"âœ”ï¸ ê¸°ì¡´ ë°ì´í„°ë² ì´ìŠ¤ íŒŒì¼ì„ ì‚¬ìš©í•©ë‹ˆë‹¤: {db_file_path}")
    print("--------------------------------\n")

    excel_file_path = select_excel_file()

    if excel_file_path:
        print(f"ğŸ‘ íŒŒì¼ ì„ íƒ ì™„ë£Œ: '{os.path.basename(excel_file_path)}'\n")

        if add_customers_from_excel_to_db(excel_file_path, db_file_path):
            if confirm_step(2, "ì—‘ì…€ íŒŒì¼ ë¶„ë¥˜ ì‘ì—…"):
                step2_output_file = process_and_classify_excel(excel_file_path, db_file_path)

                if step2_output_file:
                    if confirm_step(3, "'ì¬ë¯¼ì›' ì„œì‹ ì ìš©"):
                        step3_output_file = apply_final_formatting(step2_output_file, db_file_path)
                        if step3_output_file:
                            if confirm_step(4, "AI ì˜ˆì¸¡ê¸° ì‹¤í–‰ ë° ê²°ê³¼ ë³‘í•©"):
                                step4_output_file = step_4_run_predictor_and_process(step3_output_file)
                                if step4_output_file:
                                    if confirm_step(5, "2ì°¨ ë°ì´í„° ê°€ê³µ ë° ìµœì¢… ë¶„ë¥˜"):
                                        step_5_run_secondary_processing(step4_output_file)
                                    else:
                                        print("ğŸš« ì‚¬ìš©ìê°€ 5ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤. 4ë‹¨ê³„ ì™„ë£Œ íŒŒì¼ì´ ìµœì¢… ê²°ê³¼ì…ë‹ˆë‹¤.")
                            else:
                                print("ğŸš« ì‚¬ìš©ìê°€ 4ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤.")
                    else:
                        print("ğŸš« ì‚¬ìš©ìê°€ 3ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤.")
            else:
                print("ğŸš« ì‚¬ìš©ìê°€ 2ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤.")
    else:
        print("\níŒŒì¼ì´ ì„ íƒë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
```

```python
import pandas as pd
import sqlite3
import os
import sys
import tkinter as tk
from tkinter import filedialog, messagebox
from openpyxl.styles import Font
import shutil
import subprocess

#ë°˜ë“œì‹œ python 3.10 ë²„ì „ìœ¼ë¡œ ëŒë ¤ì£¼ì„¸ìš”. 3.13ë²„ì „ì¼ ê²½ìš° ëŒì•„ê°€ì§€ ì•ŠìŠµë‹ˆë‹¤.

# --- 1ë‹¨ê³„ í•¨ìˆ˜ ---

def select_excel_file():
    print("ğŸ“‚ ì—‘ì…€ íŒŒì¼ ì„ íƒì„ ìœ„í•œ íŒì—…ì°½ì„ ì—½ë‹ˆë‹¤...")
    root = tk.Tk(); root.withdraw()
    return filedialog.askopenfilename(
        title="ì²˜ë¦¬í•  ì—‘ì…€ íŒŒì¼ì„ ì„ íƒí•˜ì„¸ìš”",
        filetypes=[("Excel files", "*.xlsx *.xls"), ("All files", "*.*")]
    )


def add_customers_from_excel_to_db(excel_path, db_path):
    print("âœ… 1ë‹¨ê³„: ì—‘ì…€ ë°ì´í„°ë¥¼ ë°ì´í„°ë² ì´ìŠ¤ì— ì¶”ê°€í•˜ëŠ” ì‘ì—…ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    conn = None
    try:
        conn = sqlite3.connect(db_path)
        cur = conn.cursor()
        print(f"ğŸ”— ë°ì´í„°ë² ì´ìŠ¤ '{os.path.basename(db_path)}'ì— ì—°ê²°ë˜ì—ˆìŠµë‹ˆë‹¤.")
        cur.execute('''
            CREATE TABLE IF NOT EXISTS customers (
                receipt_number TEXT PRIMARY KEY, customer_name TEXT, phone_number TEXT
            )
        ''')
        print("ğŸ“œ 'customers' í…Œì´ë¸”ì„ í™•ì¸í•˜ê³ , ì—†ìœ¼ë©´ ìƒì„±í•˜ê² ìŠµë‹ˆë‹¤.")

        df = pd.read_excel(excel_path, usecols=['ê³ ê°ëª…', 'íœ´ëŒ€í°', 'ì ‘ìˆ˜ë²ˆí˜¸'], dtype=str)
        print(f"ğŸ“„ '{os.path.basename(excel_path)}' íŒŒì¼ì—ì„œ {len(df)}ê°œì˜ ë°ì´í„°ë¥¼ ì½ì—ˆìŠµë‹ˆë‹¤.")

        original_count = len(df)
        df.dropna(subset=['ì ‘ìˆ˜ë²ˆí˜¸'], inplace=True)
        df = df[df['ì ‘ìˆ˜ë²ˆí˜¸'].astype(str).str.strip() != '']
        dropped_count = original_count - len(df)
        if dropped_count > 0:
            print(f"ğŸ§¹ 'ì ‘ìˆ˜ë²ˆí˜¸'ê°€ ë¹„ì–´ìˆëŠ” {dropped_count}ê°œì˜ í–‰ì„ DB ì¶”ê°€ ëŒ€ìƒì—ì„œ ì œì™¸í–ˆìŠµë‹ˆë‹¤.")

        new_data_count = 0
        for _, row in df.iterrows():
            # DBì— ì¶”ê°€ (ì´ë¯¸ ìˆëŠ” ì ‘ìˆ˜ë²ˆí˜¸ëŠ” ë¬´ì‹œë¨)
            cur.execute('INSERT OR IGNORE INTO customers VALUES (?, ?, ?)', (row['ì ‘ìˆ˜ë²ˆí˜¸'], row['ê³ ê°ëª…'], row['íœ´ëŒ€í°']))
            if cur.rowcount > 0:
                new_data_count += 1
        conn.commit()
        print(f"ğŸ’¾ ë°ì´í„°ë² ì´ìŠ¤ì— ì´ {new_data_count}ê°œì˜ ì‹ ê·œ ê³ ê° ì •ë³´ê°€ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return True
    except Exception as e:
        print(f"âŒ 1ë‹¨ê³„ ì˜¤ë¥˜: {e}"); root = tk.Tk(); root.withdraw(); messagebox.showerror("1ë‹¨ê³„ ì˜¤ë¥˜", f"ì—‘ì…€ ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜:\n\n{e}")
        return False
    finally:
        if conn: conn.close()
        print("ğŸ”— ë°ì´í„°ë² ì´ìŠ¤ ì—°ê²°ì„ ì¢…ë£Œí–ˆìŠµë‹ˆë‹¤.")
        print("âœ… 1ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

# --- 2ë‹¨ê³„ ê´€ë ¨ í•¨ìˆ˜ ---
def process_and_classify_excel(excel_path, db_path):
    print(f"âœ… 2ë‹¨ê³„: '{os.path.basename(excel_path)}' íŒŒì¼ ë¶„ë¥˜ ì‘ì—…ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    output_filename = None
    try:
        df = pd.read_excel(excel_path, dtype=str).fillna('')
        phone_col = 'íœ´ëŒ€í°' if 'íœ´ëŒ€í°' in df.columns else 'ì „í™”ë²ˆí˜¸'
        if phone_col not in df.columns: raise KeyError("'íœ´ëŒ€í°' ë˜ëŠ” 'ì „í™”ë²ˆí˜¸' ì—´ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

        conn = sqlite3.connect(db_path)
        cur = conn.cursor()
        print(f"ğŸ” DBì— ì—°ê²°í•˜ì—¬ ê° í–‰ì˜ ê³¼ê±° ì´ë ¥ì„ í™•ì¸í•©ë‹ˆë‹¤.")

        df['ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] = ''
        re_customer_count = 0
        unanswered_count = 0

        for index, row in df.iterrows():
            cur.execute('SELECT 1 FROM customers WHERE customer_name = ? AND phone_number = ? AND receipt_number != ? LIMIT 1',
                        (row['ê³ ê°ëª…'], row[phone_col], row['ì ‘ìˆ˜ë²ˆí˜¸']))
            is_re_complaint = cur.fetchone() is not None

            if not row['ë‹µë³€']:
                df.at[index, 'ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] = 'ë¯¸ë‹µë³€'; unanswered_count += 1
            if is_re_complaint:
                df.at[index, 'ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] = 'ì¬ë¯¼ì›'; re_customer_count += 1
                if not row['ë‹µë³€']: unanswered_count -= 1

        conn.close()
        print(f"ğŸ“Š ë¶„ë¥˜ ì™„ë£Œ: ì¬ë¯¼ì› {re_customer_count}ê±´, ë¯¸ë‹µë³€ {unanswered_count}ê±´")

        base, ext = os.path.splitext(excel_path)
        output_filename = f"{base}_2ë‹¨ê³„ì™„ë£Œ{ext}"
        df.to_excel(output_filename, index=False)
        print(f"ğŸ’¾ ê²°ê³¼ê°€ '{os.path.basename(output_filename)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return output_filename
    except Exception as e:
        print(f"âŒ 2ë‹¨ê³„ ì˜¤ë¥˜: {e}"); root = tk.Tk(); root.withdraw(); messagebox.showerror("2ë‹¨ê³„ ì˜¤ë¥˜", f"ë¶„ë¥˜ ì‘ì—… ì¤‘ ì˜¤ë¥˜:\n\n{e}")
        return None
    finally:
        print("âœ… 2ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

# --- 3ë‹¨ê³„ í•¨ìˆ˜ ---
def apply_final_formatting(input_path, db_path):
    print(f"âœ… 3ë‹¨ê³„: '{os.path.basename(input_path)}' íŒŒì¼ì— ìµœì¢… ì„œì‹ ì ìš©ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    try:
        df = pd.read_excel(input_path, dtype=str).fillna('')

        if 'ë“±ë¡ì¼' in df.columns:
            df['ë“±ë¡ì¼'] = pd.to_datetime(df['ë“±ë¡ì¼'], errors='coerce')
            print("ğŸ“… 'ë“±ë¡ì¼' ì—´ì„ ë‚ ì§œ í˜•ì‹ìœ¼ë¡œ ë³€í™˜í–ˆìŠµë‹ˆë‹¤.")
        else:
            print("âš ï¸ 'ë“±ë¡ì¼' ì—´ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ì²« ë²ˆì§¸ ë¯¼ì› ì œì™¸ ë¡œì§ì„ ê±´ë„ˆëœë‹ˆë‹¤.")

        re_complaint_mask = (df['ë‚´ìš©ë¶„ë¥˜_ì¢…í•©_ì˜ˆì¸¡'] == 'ì¬ë¯¼ì›')

        if 'ë“±ë¡ì¼' in df.columns and not df['ë“±ë¡ì¼'].isna().all():
            df['ê³ ê°í‚¤'] = df['ê³ ê°ëª…'] + '_' + df['íœ´ëŒ€í°']
            # í˜„ì¬ íŒŒì¼ ê¸°ì¤€, ê³ ê°ë³„ ì²« ë¯¼ì›ì˜ ì¸ë±ìŠ¤ë¥¼ ì°¾ìŒ
            first_complaint_indices = df.groupby('ê³ ê°í‚¤')['ë“±ë¡ì¼'].idxmin()

            indices_to_exclude = []
            conn = None
            try:
                print("ğŸ” ë°ì´í„°ë² ì´ìŠ¤ë¥¼ í™•ì¸í•˜ì—¬ 'ì§„ì§œ' ì²« ë¯¼ì› ì—¬ë¶€ë¥¼ ê²€ì¦í•©ë‹ˆë‹¤...")
                conn = sqlite3.connect(db_path)
                cur = conn.cursor()

                # í˜„ì¬ íŒŒì¼ì˜ ì²« ë¯¼ì› í›„ë³´ë“¤ì— ëŒ€í•´ DB ì´ë ¥ì„ í•˜ë‚˜ì”© í™•ì¸
                first_complaint_df = df.loc[first_complaint_indices]
                for idx, row in first_complaint_df.iterrows():
                    cur.execute('SELECT COUNT(*) FROM customers WHERE customer_name = ? AND phone_number = ?', (row['ê³ ê°ëª…'], row['íœ´ëŒ€í°']))
                    db_count = cur.fetchone()[0]

                    # DB ì´ë ¥ì´ 2ê±´ ë¯¸ë§Œì¸ ê³ ê°ë§Œ 'ì²« ë¯¼ì› ì œì™¸' ê·œì¹™ì„ ì ìš©
                    if db_count < 2:
                        indices_to_exclude.append(idx)

                print(f"ğŸ¯ DB ì´ë ¥ ê²€ì¦ ì™„ë£Œ: 'ì²« ë¯¼ì›' í›„ë³´ {len(first_complaint_indices)}ê±´ ì¤‘ {len(indices_to_exclude)}ê±´ì„ ì¬ë¯¼ì›ì—ì„œ ìµœì¢… ì œì™¸í•©ë‹ˆë‹¤.")

            except Exception as e:
                print(f"âš ï¸ DB ê²€ì¦ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}. ê¸°ì¡´ ë°©ì‹(íŒŒì¼ ê¸°ì¤€)ìœ¼ë¡œ ì²« ë¯¼ì› ì œì™¸ ë¡œì§ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.")
                indices_to_exclude = list(first_complaint_indices)
            finally:
                if conn: conn.close()

            re_complaint_mask = re_complaint_mask & ~df.index.isin(indices_to_exclude)

        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜1'] = '(ë¬¼ì )ê¸°íƒ€'
        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜2'] = 'ê¸°íƒ€'
        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜3'] = 'ê¸°íƒ€'
        df.loc[re_complaint_mask, 'ë‚´ìš©ë¶„ë¥˜4'] = 'ì¬ë¯¼ì›'
        print("âœï¸ 'ì¬ë¯¼ì›' í•­ëª©ì— ëŒ€í•œ í…ìŠ¤íŠ¸ ì±„ìš°ê¸°ë¥¼ ì™„ë£Œí–ˆìŠµë‹ˆë‹¤.")

        base, ext = os.path.splitext(input_path)
        base = base.replace('_2ë‹¨ê³„ì™„ë£Œ', '')
        output_filename = f"{base}_3ë‹¨ê³„ì™„ë£Œ{ext}"

        with pd.ExcelWriter(output_filename, engine='openpyxl') as writer:
            df.drop(columns=['ê³ ê°í‚¤'], inplace=True, errors='ignore')
            df.to_excel(writer, index=False, sheet_name='Sheet1')
            worksheet = writer.sheets['Sheet1']
            red_font = Font(color="FF0000")
            target_cols = ['ë‚´ìš©ë¶„ë¥˜1', 'ë‚´ìš©ë¶„ë¥˜2', 'ë‚´ìš©ë¶„ë¥˜3', 'ë‚´ìš©ë¶„ë¥˜4']
            col_indices = {col: list(df.columns).index(col) + 1 for col in target_cols if col in df.columns}

            for row_idx in df.index[re_complaint_mask]:
                excel_row = row_idx + 2
                for col_num in col_indices.values():
                    worksheet.cell(row=excel_row, column=col_num).font = red_font

        print("ğŸ¨ 'ì¬ë¯¼ì›' í•­ëª©ì— ëŒ€í•œ ê¸€ììƒ‰ ì„œì‹ ì ìš©ì„ ì™„ë£Œí–ˆìŠµë‹ˆë‹¤.")
        full_path = os.path.abspath(output_filename)
        print(f"ğŸ’¾ 3ë‹¨ê³„ ê²°ê³¼ê°€ '{os.path.basename(output_filename)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return full_path
    except Exception as e:
        print(f"âŒ 3ë‹¨ê³„ ì˜¤ë¥˜: {e}"); root = tk.Tk(); root.withdraw(); messagebox.showerror("3ë‹¨ê³„ ì˜¤ë¥˜", f"ì„œì‹ ì ìš© ì¤‘ ì˜¤ë¥˜:\n\n{e}")
        return None
    finally:
        print("âœ… 3ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

def step_4_run_predictor_and_process(step3_output_path):
    print(f"âœ… 4ë‹¨ê³„: ì™¸ë¶€ ì˜ˆì¸¡ ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰ ë° ê²°ê³¼ ë³‘í•©ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    final_output_path = None
    predictor_input_path = None
    predictor_output_path = None

    try:
        script_dir = os.path.dirname(os.path.abspath(__file__))
        predictor_script_name = 'predictor_for_categorize.py'
        script_to_run_path = os.path.join(script_dir, 'module', predictor_script_name)

        predictor_input_filename = 'VOC_ëª©ë¡.xlsx'
        predictor_output_filename = 'predicted_data.xlsx'
        predictor_input_path = os.path.join(script_dir, predictor_input_filename)
        predictor_output_path = os.path.join(script_dir, predictor_output_filename)

        shutil.copy(step3_output_path, predictor_input_path)
        print(f"ğŸ“‹ '{os.path.basename(step3_output_path)}'ë¥¼ '{predictor_input_filename}'ìœ¼ë¡œ ë³µì‚¬í–ˆìŠµë‹ˆë‹¤.")

        if not os.path.exists(script_to_run_path):
            raise FileNotFoundError(f"ì‹¤í–‰í•  ìŠ¤í¬ë¦½íŠ¸ '{script_to_run_path}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤!")

        print(f"ğŸš€ '{predictor_script_name}' ìŠ¤í¬ë¦½íŠ¸ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤...")
        env = os.environ.copy()
        env.update({'TF_CPP_MIN_LOG_LEVEL': '2', 'TF_ENABLE_ONEDNN_OPTS': '0'})

        process = subprocess.Popen(
            [sys.executable, "-u", script_to_run_path],
            stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
            cwd=script_dir, env=env
        )

        for line_bytes in process.stdout:
            try:
                line = line_bytes.decode('utf-8')
            except UnicodeDecodeError:
                line = line_bytes.decode('cp949', errors='replace') # í˜¹ì‹œ ëª¨ë¥¼ ì˜¤ë¥˜ ë°©ì§€
            print(line, end='', flush=True)

        process.wait()
        if process.returncode != 0:
            raise subprocess.CalledProcessError(process.returncode, [sys.executable, script_to_run_path])

        print(f"\nğŸ‰ '{predictor_script_name}' ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰ ì™„ë£Œ!")

        if not os.path.exists(predictor_output_path):
            raise FileNotFoundError(f"ì˜ˆì¸¡ ìŠ¤í¬ë¦½íŠ¸ê°€ ê²°ê³¼ íŒŒì¼('{predictor_output_filename}')ì„ ìƒì„±í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤!")

        print(f"ğŸ“Š ì›ë³¸ íŒŒì¼ê³¼ ì˜ˆì¸¡ ê²°ê³¼ íŒŒì¼ì„ ì½ìŠµë‹ˆë‹¤.")
        df_original = pd.read_excel(step3_output_path, dtype=str).fillna('')
        df_predicted = pd.read_excel(predictor_output_path, dtype=str).fillna('')

        print("âœï¸ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ì›ë³¸ ë°ì´í„°ì— ë³‘í•©í•˜ê³  ê°€ê³µí•©ë‹ˆë‹¤...")
        df_original['ë‚´ìš©ë¶„ë¥˜'] = df_predicted.get('ë‚´ìš©ë¶„ë¥˜', '')
        df_original['ë‚´ìš©ë¶„ë¥˜(2ì°¨)'] = df_predicted.get('ë‚´ìš©ë¶„ë¥˜(2ì°¨)', '')

        re_complaint_mask = (df_original['ë‚´ìš©ë¶„ë¥˜4'] == 'ì¬ë¯¼ì›')

        for index, row in df_original.iterrows():
            if re_complaint_mask[index]: continue

            primary_cat, secondary_cat = str(row.get('ë‚´ìš©ë¶„ë¥˜', '')), str(row.get('ë‚´ìš©ë¶„ë¥˜(2ì°¨)', ''))
            source_string = None

            if primary_cat and primary_cat != 'ê¸°íƒ€': source_string = primary_cat
            elif primary_cat == 'ê¸°íƒ€' and secondary_cat: source_string = secondary_cat

            if source_string:
                parts = [p.strip() for p in source_string.split(' > ')]
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜1'] = parts[0] if len(parts) > 0 else ''
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜2'] = parts[1] if len(parts) > 1 else ''
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜3'] = parts[2] if len(parts) > 2 else ''
                df_original.loc[index, 'ë‚´ìš©ë¶„ë¥˜4'] = parts[3] if len(parts) > 3 else ''

        print("ë³‘í•© ë° ê°€ê³µ ì™„ë£Œ.")

        base, ext = os.path.splitext(step3_output_path)
        final_output_path = base.replace('_3ë‹¨ê³„ì™„ë£Œ', '_4ë‹¨ê³„ì™„ë£Œ') + ext
        df_original.drop(columns=['ë‚´ìš©ë¶„ë¥˜', 'ë‚´ìš©ë¶„ë¥˜(2ì°¨)'], inplace=True, errors='ignore')
        df_original.to_excel(final_output_path, index=False)

        print(f"ğŸ’¾ 4ë‹¨ê³„ ê²°ê³¼ê°€ '{os.path.basename(final_output_path)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
        return final_output_path

    except (FileNotFoundError, subprocess.CalledProcessError, Exception) as e:
        error_type = type(e).__name__
        print(f"âŒ 4ë‹¨ê³„ ì˜¤ë¥˜ ({error_type}): {e}")
        root = tk.Tk(); root.withdraw(); messagebox.showerror(f"4ë‹¨ê³„ {error_type}", f"ì˜¤ë¥˜ ë°œìƒ:\n\n{e}")
        return None
    finally:
        if predictor_input_path and os.path.exists(predictor_input_path):
            os.remove(predictor_input_path); print(f"ğŸ§¹ ì„ì‹œ ì…ë ¥ íŒŒì¼ ì‚­ì œ ì™„ë£Œ.")
        if predictor_output_path and os.path.exists(predictor_output_path):
            os.remove(predictor_output_path); print(f"ğŸ§¹ ì„ì‹œ ê²°ê³¼ íŒŒì¼ ì‚­ì œ ì™„ë£Œ.")
        print("âœ… 4ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")


def step_5_run_secondary_processing(step4_output_path):
    print(f"âœ… 5ë‹¨ê³„: 2ì°¨ ë°ì´í„° ê°€ê³µ ìŠ¤í¬ë¦½íŠ¸ ì‹¤í–‰ì„ ì‹œì‘í•©ë‹ˆë‹¤.")
    final_output_path = None

    try:
        script_dir = os.path.dirname(os.path.abspath(__file__))
        script_name = 'second_process.py'
        script_to_run_path = os.path.join(script_dir, 'module', script_name)

        if not os.path.exists(script_to_run_path):
            raise FileNotFoundError(f"ì‹¤í–‰í•  ìŠ¤í¬ë¦½íŠ¸ '{script_to_run_path}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤!")

        base, ext = os.path.splitext(step4_output_path)
        final_output_path = base.replace('_4ë‹¨ê³„ì™„ë£Œ', '_5ë‹¨ê³„ì™„ë£Œ_ìµœì¢…') + ext

        print(f"ğŸš€ '{script_name}' ìŠ¤í¬ë¦½íŠ¸ë¥¼ ì‹¤í–‰í•©ë‹ˆë‹¤...")

        process = subprocess.Popen(
            [sys.executable, "-u", script_to_run_path, step4_output_path, final_output_path],
            stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
            cwd=script_dir
        )

        for line_bytes in process.stdout:
            try:
                line = line_bytes.decode('utf-8')
            except UnicodeDecodeError:
                line = line_bytes.decode('cp949', errors='replace')
            print(line, end='', flush=True)

        process.wait()
        if process.returncode != 0:
            raise subprocess.CalledProcessError(process.returncode, [sys.executable, script_to_run_path])

        if not os.path.exists(final_output_path):
            raise FileNotFoundError("2ì°¨ ê°€ê³µ ìŠ¤í¬ë¦½íŠ¸ê°€ ê²°ê³¼ íŒŒì¼ì„ ìƒì„±í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤!")

        full_path = os.path.abspath(final_output_path)
        success_msg = f"ëª¨ë“  ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!\n\nìµœì¢… ê²°ê³¼ê°€ '{os.path.basename(final_output_path)}' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.\n\nì €ì¥ ìœ„ì¹˜:\n{full_path}"
        print(f"\nğŸ’¾ {success_msg}")
        root = tk.Tk(); root.withdraw(); messagebox.showinfo("ìµœì¢… ì™„ë£Œ", success_msg)
        return final_output_path

    except (FileNotFoundError, subprocess.CalledProcessError, Exception) as e:
        error_type = type(e).__name__
        print(f"âŒ 5ë‹¨ê³„ ì˜¤ë¥˜ ({error_type}): {e}")
        root = tk.Tk(); root.withdraw(); messagebox.showerror(f"5ë‹¨ê³„ {error_type}", f"ì˜¤ë¥˜ ë°œìƒ:\n\n{e}")
        return None
    finally:
        print("âœ… 5ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")


def confirm_step(step_number, description):
    print(f"ğŸ¤” {step_number}ë‹¨ê³„ ì§„í–‰ ì—¬ë¶€ë¥¼ ë¬»ëŠ” íŒì—…ì°½ì„ ì—½ë‹ˆë‹¤...")
    root = tk.Tk(); root.withdraw()
    root.attributes('-topmost', True)
    return messagebox.askyesno(
        title=f"{step_number}ë‹¨ê³„ ì§„í–‰ í™•ì¸",
        message=f"{step_number-1}ë‹¨ê³„ ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n\nì´ì–´ì„œ {step_number}ë‹¨ê³„({description})ë¥¼ ì§„í–‰í•˜ì‹œê² ìŠµë‹ˆê¹Œ?"
    )

# --- ë©”ì¸  ---
if __name__ == "__main__":
    try:
        script_dir = os.path.dirname(os.path.abspath(__file__))
    except NameError:
        script_dir = os.getcwd()
    db_file_name = 'ëˆ„ì _ê³ ê°_2025.db'
    db_file_path = os.path.join(script_dir, db_file_name)

    print("--- ğŸ“‚ ë°ì´í„°ë² ì´ìŠ¤ íŒŒì¼ í™•ì¸ ---")
    if not os.path.exists(db_file_path):
        print(f"â„¹ï¸ '{db_file_name}' íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ìƒˆ íŒŒì¼ì„ ìƒì„±í•©ë‹ˆë‹¤.")
    else:
        print(f"âœ”ï¸ ê¸°ì¡´ ë°ì´í„°ë² ì´ìŠ¤ íŒŒì¼ì„ ì‚¬ìš©í•©ë‹ˆë‹¤: {db_file_path}")
    print("--------------------------------\n")

    excel_file_path = select_excel_file()

    if excel_file_path:
        print(f"ğŸ‘ íŒŒì¼ ì„ íƒ ì™„ë£Œ: '{os.path.basename(excel_file_path)}'\n")

        if add_customers_from_excel_to_db(excel_file_path, db_file_path):
            if confirm_step(2, "ì—‘ì…€ íŒŒì¼ ë¶„ë¥˜ ì‘ì—…"):
                step2_output_file = process_and_classify_excel(excel_file_path, db_file_path)

                if step2_output_file:
                    if confirm_step(3, "'ì¬ë¯¼ì›' ì„œì‹ ì ìš©"):
                        step3_output_file = apply_final_formatting(step2_output_file, db_file_path)
                        if step3_output_file:
                            if confirm_step(4, "AI ì˜ˆì¸¡ê¸° ì‹¤í–‰ ë° ê²°ê³¼ ë³‘í•©"):
                                step4_output_file = step_4_run_predictor_and_process(step3_output_file)
                                if step4_output_file:
                                    if confirm_step(5, "2ì°¨ ë°ì´í„° ê°€ê³µ ë° ìµœì¢… ë¶„ë¥˜"):
                                        step_5_run_secondary_processing(step4_output_file)
                                    else:
                                        print("ğŸš« ì‚¬ìš©ìê°€ 5ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤. 4ë‹¨ê³„ ì™„ë£Œ íŒŒì¼ì´ ìµœì¢… ê²°ê³¼ì…ë‹ˆë‹¤.")
                            else:
                                print("ğŸš« ì‚¬ìš©ìê°€ 4ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤.")
                    else:
                        print("ğŸš« ì‚¬ìš©ìê°€ 3ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤.")
            else:
                print("ğŸš« ì‚¬ìš©ìê°€ 2ë‹¨ê³„ ì§„í–‰ì„ ì·¨ì†Œí–ˆìŠµë‹ˆë‹¤.")
    else:
        print("\níŒŒì¼ì´ ì„ íƒë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
```

```python
import pandas as pd
import json
import torch
import torch.nn.functional as F
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
import tkinter as tk
from tkinter import filedialog
import os
import sys

# ìœˆë„ìš° ì¸ì½”ë”© ì—ëŸ¬ ë°©ì§€ ì„¤ì • (í•„ìš”ì‹œ ì‚¬ìš©)
# sys.stdout.reconfigure(encoding='utf-8')

# ==========================================
# 1. ì„¤ì • ë° ìµœì í™” í´ë˜ìŠ¤ ì •ì˜
# ==========================================

class TextDataset(Dataset):
    """ë°ì´í„°ë¥¼ íš¨ìœ¨ì ìœ¼ë¡œ ë°°ì¹˜ ë‹¨ìœ„ë¡œ ê³µê¸‰í•˜ê¸° ìœ„í•œ í´ë˜ìŠ¤"""
    def __init__(self, texts):
        self.texts = texts

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return str(self.texts[idx])

def load_optimized_model(model_path):
    print("[INFO] ëª¨ë¸ ë¡œë”© ë° ìµœì í™” ì¤‘... (ì ì‹œë§Œ ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”)", flush=True)

    # 1. í† í¬ë‚˜ì´ì € ë¡œë“œ
    tokenizer = AutoTokenizer.from_pretrained(model_path)

    # 2. ëª¨ë¸ ë¡œë“œ
    model = AutoModelForSequenceClassification.from_pretrained(model_path, use_safetensors=True)

    # 3. [í•µì‹¬] ë™ì  ì–‘ìí™” (Quantization) ì ìš©
    # float32 -> qint8ë¡œ ë³€í™˜í•˜ì—¬ CPU ì—°ì‚° ì†ë„ ê·¹ëŒ€í™”
    print("[INFO] CPU ìµœì í™”(ì–‘ìí™”) ì ìš© ì¤‘...", flush=True)
    quantized_model = torch.quantization.quantize_dynamic(
        model,
        {torch.nn.Linear},  # ì„ í˜• ë ˆì´ì–´ë¥¼ ì–‘ìí™”
        dtype=torch.qint8
    )

    device = torch.device("cpu")
    quantized_model.to(device)
    quantized_model.eval()

    return quantized_model, tokenizer, device

def predict_optimized(df, model, tokenizer, label_mapping, device, batch_size=32):
    """ìµœì í™”ëœ ì˜ˆì¸¡ í•¨ìˆ˜ (DataLoader + Dynamic Padding ì‚¬ìš©)"""

    # ë¹ˆ ê°’ ì²˜ë¦¬
    responses = df['ë‹µë³€'].fillna('').astype(str).tolist()

    # ë°ì´í„°ì…‹ ë° ë¡œë” ìƒì„±
    dataset = TextDataset(responses)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    predictions_1st = []
    predictions_2nd = []

    print(f"\n[START] ê³ ì† ì˜ˆì¸¡ ì‹œì‘ (ì´ {len(responses)}ê±´, ë°°ì¹˜í¬ê¸° {batch_size})")

    # tqdm ì‚¬ìš© ì‹œ ascii=True ì˜µì…˜ ì¶”ê°€ (íŠ¹ìˆ˜ë¬¸ì ì—ëŸ¬ ë°©ì§€)
    with torch.no_grad():
        for batch_texts in tqdm(dataloader, desc="Analyzing", unit="batch", ascii=True):
            # [í•µì‹¬] Dynamic Padding
            inputs = tokenizer(
                batch_texts,
                padding=True,       # ë°°ì¹˜ ë‚´ ìµœì¥ ê¸¸ì´ë¡œ íŒ¨ë”©
                truncation=True,    # ëª¨ë¸ ìµœëŒ€ ê¸¸ì´(ë³´í†µ 512)ëŠ” ë„˜ì§€ ì•Šê²Œ ìë¦„
                max_length=512,
                return_tensors="pt"
            ).to(device)

            outputs = model(**inputs)
            logits = outputs.logits
            probs = F.softmax(logits, dim=-1).numpy() # CPU ì–‘ìí™” ëª¨ë¸ì€ ë°”ë¡œ numpy ë³€í™˜ ê°€ëŠ¥

            # ë°°ì¹˜ ë‚´ ê²°ê³¼ ì²˜ë¦¬
            for p in probs:
                # ìƒìœ„ 2ê°œ ì¶”ì¶œ
                top_indices = p.argsort()[-2:][::-1] # ìƒìœ„ 2ê°œ ì¸ë±ìŠ¤
                top_probs = p[top_indices]

                # 1ìˆœìœ„
                pred_1 = label_mapping.get(top_indices[0], "Unknown")
                predictions_1st.append(pred_1)

                # 2ìˆœìœ„ (í™•ë¥  30% ë¯¸ë§Œì¼ ë•Œë§Œ)
                if top_probs[0] < 0.30:
                    pred_2 = label_mapping.get(top_indices[1], "Unknown")
                    predictions_2nd.append(pred_2)
                else:
                    predictions_2nd.append("")

    df['ë‚´ìš©ë¶„ë¥˜'] = predictions_1st
    df['ë‚´ìš©ë¶„ë¥˜(2ì°¨)'] = predictions_2nd

    return df

def select_file(title_text):
    root = tk.Tk()
    root.withdraw()
    root.attributes('-topmost', True)
    file_path = filedialog.askopenfilename(title=title_text, filetypes=[("Excel files", "*.xlsx *.xls")])
    root.destroy()
    return file_path

# ==========================================
# 2. ë©”ì¸ ì‹¤í–‰ ë¡œì§
# ==========================================

if __name__ == "__main__":
    # ê²½ë¡œ ì„¤ì •
    base_path = "C:/Users/Korail/Documents/OEJ/vocë‚´ìš©ë¶„ë¥˜"
    model_dir = os.path.join(base_path, "model5")
    mapping_path = os.path.join(model_dir, "label_mapping.json")

    try:
        # 1. ë¼ë²¨ ë§¤í•‘ ë¡œë“œ
        with open(mapping_path, "r", encoding='utf-8') as f:
            label_mapping = {int(k): v for k, v in json.load(f).items()}
        print("[OK] ë¼ë²¨ ë§¤í•‘ ë¡œë“œ ì™„ë£Œ")

        # 2. ëª¨ë¸ ë¡œë“œ (ìµœì í™” ì ìš©)
        model, tokenizer, device = load_optimized_model(model_dir)
        print("[OK] ëª¨ë¸ ìµœì í™” ë° ë¡œë“œ ì™„ë£Œ")

        # 3. ë¶„ì„í•  íŒŒì¼ ì„ íƒ (ìë™ ì§€ì •)
        target_filename = "VOC_ëª©ë¡.xlsx"

        # í˜„ì¬ ì‹¤í–‰ ì¤‘ì¸ í´ë”(getcwd)ì—ì„œ íŒŒì¼ì„ ì°¾ìŠµë‹ˆë‹¤.
        input_file_path = os.path.join(os.getcwd(), target_filename)

        print(f"\n[INPUT] íŒŒì¼ì„ ìë™ìœ¼ë¡œ ì°¾ìŠµë‹ˆë‹¤: {target_filename}")

        # íŒŒì¼ì´ ì§„ì§œ ìˆëŠ”ì§€ í™•ì¸ (ì—†ìœ¼ë©´ ì—ëŸ¬ ë‚˜ë‹ˆê¹Œìš”)
        if not os.path.exists(input_file_path):
            print(f"[ERROR] í˜„ì¬ í´ë”ì— '{target_filename}' íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤!")
            print(f" -> í˜„ì¬ ìœ„ì¹˜: {os.getcwd()}")
            exit()

        print(f"[CHECK] íŒŒì¼ í™•ì¸ë¨: {input_file_path}")

        print(f"[READ] ë°ì´í„° ì½ëŠ” ì¤‘: {os.path.basename(input_file_path)}")
        df = pd.read_excel(input_file_path)

        # 4. ì˜ˆì¸¡ ìˆ˜í–‰
        result_df = predict_optimized(df, model, tokenizer, label_mapping, device, batch_size=8)

        # 5. ì €ì¥
        input_dir = os.path.dirname(input_file_path)
        input_filename = os.path.splitext(os.path.basename(input_file_path))[0]
        output_path = os.path.join(input_dir, f"predicted_data.xlsx")

        result_df.to_excel(output_path, index=False)
        print(f"\n[DONE] ë¶„ì„ ì™„ë£Œ!")
        print(f"[SAVE] ê²°ê³¼ ì €ì¥ë¨: {output_path}")

    except Exception as e:
        print(f"\n[ERROR] ì˜¤ë¥˜ ë°œìƒ: {e}")
        import traceback
        traceback.print_exc()
```
