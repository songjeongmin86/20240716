```python
# KTX 10개 역사 2022, 2023 연간 이용객 데이터
# 형식: [역명, 연도, 연간_이용객수]
raw_data = [
    ["광명",    2022,   9834349],
    ["광명",    2023,  11718891],

    ["광주송정", 2022,   4176882],
    ["광주송정", 2023,   4862628],

    ["대전",    2022,   9762345],
    ["대전",    2023,  11598990],

    ["동대구",  2022,  11232702],
    ["동대구",  2023,  12729324],

    ["부산",    2022,  10224377],
    ["부산",    2023,  13176729],

    ["서울",    2022,  26755372],
    ["서울",    2023,  32630853],

    ["오송",    2022,   7066555],
    ["오송",    2023,   8407695],

    ["용산",    2022,   9064579],
    ["용산",    2023,  10749689],

    ["울산",    2022,   3590080],
    ["울산",    2023,   4412747],

    ["천안아산", 2022,   7456112],
    ["천안아산", 2023,   8681334],
]

print("전체 데이터 건수:", len(raw_data))
print("예시 한 줄:", raw_data[0])
print()
```

```python
# raw_data를 한 줄씩 보면서 서울역 데이터를 찾는다.
for row in raw_data:
    station = row[0]  # 역명
    year    = row[1]  # 연도
    total   = row[2]  # 연간 이용객

    if station == "서울" and year == 2022:
        seoul_2022 = total
    elif station == "서울" and year == 2023:
        seoul_2023 = total

# 증가 인원 계산
seoul_diff = seoul_2023 - seoul_2022

print("문제 1 결과")
print("서울역 2022년:", seoul_2022, "명")
print("서울역 2023년:", seoul_2023, "명")
print("서울역 증가 인원:", seoul_diff, "명")
print()
```

```python
data_2022 = []

for row in raw_data:
    station = row[0]
    year    = row[1]
    total   = row[2]

    if year == 2022:               # 2022년이면
        data_2022.append([station, total])

print("문제 2 결과: data_2022")
for row in data_2022:
    print(row)
print()
```

```python
data_2023 = []

for row in raw_data:
    station = row[0]
    year    = row[1]
    total   = row[2]

    if year == 2023:               # 2023년이면
        data_2023.append([station, total])

print("문제 3 결과: data_2023")
for row in data_2023:
    print(row)
print()
```

```python
increase_list = []

# 2022 리스트를 기준으로 돈다.
for row22 in data_2022:
    station_22 = row22[0]  # 역명
    total_22   = row22[1]  # 2022 합계

    # 같은 역의 2023 데이터를 찾기 위해 data_2023 전체를 한 번 돈다.
    for row23 in data_2023:
        station_23 = row23[0]
        total_23    = row23[1]

        if station_23 == station_22:  # 역명이 같으면
            break                     # 찾았으니 더 볼 필요 없음

    diff = total_23 - total_22        # 증가 인원 계산
    increase_list.append([station_22, total_22, total_23, diff])

print("문제 4 결과: increase_list (역별 2022→2023 증가 인원)")
for row in increase_list:
    print(row)
print()
```

```python
N = 8000000  # 최소 기준 인원 (800만 명)
candidates = []

for row in increase_list:
    station  = row[0]
    total_22 = row[1]
    total_23 = row[2]
    diff     = row[3]

    if total_23 >= N:
        candidates.append([station, total_22, total_23, diff])

print("문제 6 결과: 조건(2023년 ≥ 8,000,000명) 만족 역 목록")
for row in candidates:
    print(row)
print()
```

```python
# sorted를 사용하여 내림차순 정렬
def get_key(row):
    return row[3]

increase_sorted = sorted(candidates, key=get_key, reverse=True)
# increase_sorted = sorted(candidates, key=lambda r: r[3], reverse=True)

print(increase_sorted)
```

```python
rank = 1
for row in increase_sorted[:3]:
    station  = row[0]
    total_22 = row[1]
    total_23 = row[2]
    diff     = row[3]

    print(f"{rank}위) {station}역")
    print(f" - 2022년 이용객: {total_22:,}명")
    print(f" - 2023년 이용객: {total_23:,}명")
    print(f" - 증가 인원    : {diff:,}명\n")

    rank += 1
```
