```python
# while은 조건이 참(True)인 동안 계속 반복합니다.
# 예: 월별 관객 수 누적으로 목표 관객 수에 도달할 때까지 반복
total_audience = 0   # 현재까지 누적 관객 수
month = 0            # 지난 개월 수
monthly = 300000     # 매달 추가 관객 수(명)

target = 2000000     # 목표 관객 수

while total_audience < target:  # 목표에 도달하기 전까지 반복
    month = month + 1           # 한 달이 지났다고 기록
    total_audience = total_audience + monthly  # 관객 수를 더함
    print(month, "개월차 누적 관객 수:", total_audience)
```

```python
# 박스오피스 예시 데이터를 아주 작게 만듭니다.
movie_title = "서울의 밤"   # 영화 제목(문자열)
revenue = 5200              # 매출(만원 단위, 정수)
rating = 8.4                # 평점(실수)

# 함수 만들기: def로 시작하고, 이름() 다음에 콜론(:)을 씁니다.
def show_basic_info():
    # 이 함수는 화면에 간단한 정보를 보여줍니다. (입력값/반환값 없음)
    print("영화 제목:", movie_title)
    print("매출(만원):", revenue)
    print("평점:", rating)

# 함수 호출: 만든 함수 이름 뒤에 ()를 붙여 실행합니다.
show_basic_info()
```
