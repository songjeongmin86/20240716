```python
import pandas as pd

# CSV 읽기
df = pd.read_csv("KC_KOBIS_BOX_OFFIC_MOVIE_INFO_202412.csv", encoding="utf-8")

# 실제 열 이름 확인
print(df.columns)

# 관객수(VIEWNG_NMPR_CO)를 정수형으로 변환
df["VIEWNG_NMPR_CO"] = (
    df["VIEWNG_NMPR_CO"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.strip()
    .astype(int)
)

# 예시: 장르(GENRE_NM)별 평균 관객 수
avg = df.groupby("GENRE_NM")["VIEWNG_NMPR_CO"].mean()
print(avg)

# groupby한 열을 index로 만들지 않음.
avg_noindex = df.groupby("GENRE_NM", as_index=False)["VIEWNG_NMPR_CO"].mean()
print(avg_noindex)
```

```python
sum_value = df.groupby("DISTB_CMPNY_NM")["VIEWNG_NMPR_CO"].sum()
print(sum_value)
```

```python
avg_value = df.groupby("OPN_DE", as_index=False)["VIEWNG_NMPR_CO"].mean()
print(avg_value)
```

```python
pd.set_option('display.float_format', '{:,.0f}'.format)  # 3자리수 콤마 적용
avg_value = df.groupby("OPN_DE", as_index=False)["VIEWNG_NMPR_CO"].mean()
avg_sorted = avg_value.sort_values("VIEWNG_NMPR_CO", ascending=False)
print(avg_sorted)
```

```python
import pandas as pd

# 첫 번째 표: 영화 제목과 관객 수
info1 = pd.DataFrame({
    "MOVIE_NM": ["소방관", "하얼빈", "모아나 2"],
    "VIEWNG_NMPR_CO": [3310341, 2500000, 2441853]
})

# 두 번째 표: 영화 제목과 장르
info2 = pd.DataFrame({
    "MOVIE_NM": ["소방관", "모아나 2", "위키드"],
    "GENRE_NM": ["드라마", "애니메이션", "판타지"]
})

print("info1:")
print(info1)
print("info2:")
print(info2)

# 기본 merge (inner join, 교집합)
df_inner = pd.merge(info1, info2, on="MOVIE_NM", how="inner")
print("\n[inner 결합 결과]")
print(df_inner)
```

```python
# 1) left join: 왼쪽(info1) 기준
df_left = pd.merge(info1, info2, on="MOVIE_NM", how="left")
print("\n[left 결합 결과]")
print(df_left)
```

```python
# 2) right join: 오른쪽(info2) 기준
df_right = pd.merge(info1, info2, on="MOVIE_NM", how="right")
print("\n[right 결합 결과]")
print(df_right)
```

```python
# 3) outer join: 전체(합집합)
df_outer = pd.merge(info1, info2, on="MOVIE_NM", how="outer")
print("\n[outer 결합 결과]")
print(df_outer)
```
