```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split # 데이터나누기, 4(train:연습문제(문제, 정답), test:수능문제(문제, 정답))
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/광주광역시2024년아파트(전월세)_실거래가_rf.csv', sep='\t')
```

```python
df.head()
```

```python
df.info()
```

```python
price_median = np.median(df['보증금'])
price_median
```

```python
y = df['보증금']
X = df.drop(columns = ['보증금'])
```

```python
X.head()
```

```python
df.corr()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=777)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f'R² Score: {r2:.4f}')
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse:.4f}')
```
