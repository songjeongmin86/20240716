```python
from requests import get
import json
import pandas as pd
from datetime import datetime
import requests # Added for more specific exception handling


def get_naverapi(query):
    product_lists = []
    elements = [] # Initialize elements to an empty list to prevent UnboundLocalError

    timestamp = datetime.now().strftime('%Y%m%d%H%M%S')
    client_id = "J2wW6Xu6fxJiS0zTlCD0"
    client_secret = "SYJBdwzack"

    user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64) "\
        "AppleWebKit/537.36 (KHTML, like Gecko) " \
        "Chrome/51.0.2704.103 Safari/537.36"
    headers = {"User-Agent": user_agent,
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret}

    url = f"https://openapi.naver.com/v1/search/shop?query={query}&display=100"

    try:
        response = get(url, headers=headers)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
        response_data = json.loads(response.text)
        elements = response_data.get('items', []) # Use .get() to safely access 'items' or default to empty list

        if not elements:
            print(f"[경고] 검색 결과에 'items'가 없거나 비어 있습니다. 응답: {response.text}")

    except requests.exceptions.RequestException as e: # Catch all requests-related errors
        print(f"[오류] API 요청 중 오류 발생: {e}")
    except json.JSONDecodeError as e: # Catch JSON parsing errors
        print(f"[오류] JSON 디코딩 오류: {e}, 응답 텍스트: {response.text}")
    except Exception as e: # Catch any other unexpected errors
        print(f"[오류] 예기치 않은 오류 발생: {type(e).__name__}: {e}")

    for i, elm in enumerate(elements):
        title = elm['title'].replace('<b>','').replace('</b>','')
        link = elm['link']
        thumbnail = elm['image']
        lprice = elm['lprice']
        mall_name = elm['mallName']
        product_lists.append([i+1, query, title, link, thumbnail, lprice, mall_name])
        print(([i+1, query, title, link, thumbnail, lprice, mall_name]))

    # Check if product_lists is empty before creating DataFrame to avoid errors
    if not product_lists:
        print("[정보] 검색된 상품이 없어 DataFrame을 생성하지 않습니다.")
        return pd.DataFrame(columns=['no', 'query', 'title', 'link', 'thumbnail', 'lprice', 'mall_name'])

    df = pd.DataFrame(product_lists, columns=['no', 'query', 'title', 'link', 'thumbnail', 'lprice', 'mall_name'])
    df.to_csv(f'navershopping_product_lists.csv', index=False)
    df.to_csv(f'backup/navershopping_product_lists_{timestamp}.csv', index=False)

    return df


def main():
    try:
        query = "케이팝굿즈"
        get_naverapi(query)
    except Exception as e:
        print(f"[오류] {type(e).__name__}: {e}")

    return


if __name__ == "__main__":
    main()
```

### 1. 필요한 라이브러리 임포트

### 2. `get_naverapi(query)` 함수: 네이버 쇼핑 API 호출 및 데이터 처리

```python
# `def get_naverapi(query):`: 검색어를 인자로 받아 API를 호출하고 결과를 반환하는 함수를 정의합니다.
# `    product_lists = []`: 검색된 상품 정보를 저장할 빈 리스트를 초기화합니다.
# `    elements = []`: API 응답에서 추출한 상품 아이템들을 저장할 빈 리스트를 초기화합니다. (UnboundLocalError 방지)

# `    timestamp = datetime.now().strftime('%Y%m%d%H%M%S')`: 현재 시간을 'YYYYMMDDHHMMSS' 형식으로 포맷하여 타임스탬프를 생성합니다.
# `    client_id = "J2wW6Xu6fxJiS0zTlCD0"`: 네이버 개발자 센터에서 발급받은 클라이언트 ID를 설정합니다.
# `    client_secret = "SYJBdwzack"`: 네이버 개발자 센터에서 발급받은 클라이언트 시크릿을 설정합니다.

# `    user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64) "\
#         "AppleWebKit/537.36 (KHTML, like Gecko) " \
#         "Chrome/51.0.2704.103 Safari/537.36"`: API 요청 시 보낼 User-Agent 헤더를 설정합니다.
# `    headers = {"User-Agent": user_agent,
#         "X-Naver-Client-Id": client_id,
#         "X-Naver-Client-Secret": client_secret}`: API 요청 헤더를 딕셔너리 형태로 구성합니다. 클라이언트 ID와 시크릿이 포함됩니다.

# `    url = f"https://openapi.naver.com/v1/search/shop?query={query}&display=100"`: 네이버 쇼핑 검색 API 엔드포인트 URL을 구성합니다. 검색어(`query`)와 한 번에 표시할 결과 수(`display=100`)를 포함합니다.

# `    try:`: API 호출 및 JSON 파싱 과정에서 발생할 수 있는 오류를 처리하기 위한 `try-except` 블록을 시작합니다.
# `        response = get(url, headers=headers)`: 구성된 URL과 헤더를 사용하여 GET 요청을 보냅니다.
# `        response.raise_for_status()`: HTTP 응답 코드가 4xx 또는 5xx일 경우 `HTTPError` 예외를 발생시킵니다. (예: 401 Unauthorized, 404 Not Found 등)
# `        response_data = json.loads(response.text)`: API 응답 본문을 JSON 형식으로 파싱하여 파이썬 딕셔너리로 변환합니다.
# `        elements = response_data.get('items', [])`: 파싱된 데이터에서 'items' 키의 값을 가져옵니다. 만약 'items' 키가 없으면 빈 리스트(`[]`)를 기본값으로 사용합니다 (`KeyError` 방지).

# `        if not elements:`: 만약 'items' 리스트가 비어 있다면 (검색 결과 없음).
# `            print(f"[경고] 검색 결과에 'items'가 없거나 비어 있습니다. 응답: {response.text}")`: 경고 메시지를 출력합니다.

# `    except requests.exceptions.RequestException as e:`: `requests` 라이브러리 관련 모든 예외(네트워크 오류, HTTP 오류 등)를 처리합니다.
# `        print(f"[오류] API 요청 중 오류 발생: {e}")`: 오류 메시지를 출력합니다.
# `    except json.JSONDecodeError as e:`: JSON 파싱 중 오류가 발생했을 때 처리합니다.
# `        print(f"[오류] JSON 디코딩 오류: {e}, 응답 텍스트: {response.text}")`: 오류 메시지와 응답 텍스트를 출력합니다.
# `    except Exception as e:`: 위에 명시되지 않은 모든 예기치 않은 오류를 처리합니다.
# `        print(f"[오류] 예기치 않은 오류 발생: {type(e).__name__}: {e}")`: 오류 유형과 메시지를 출력합니다.

# `    for i, elm in enumerate(elements):`: `elements` 리스트의 각 아이템(`elm`)과 인덱스(`i`)를 순회합니다.
# `        title = elm['title'].replace('<b>','').replace('</b>','')`: 상품 제목에서 HTML 태그(`<b>`, `</b>`)를 제거합니다.
# `        link = elm['link']`: 상품 링크를 가져옵니다.
# `        thumbnail = elm['image']`: 상품 썸네일 이미지를 가져옵니다.
# `        lprice = elm['lprice']`: 상품 최저가를 가져옵니다.
# `        mall_name = elm['mallName']`: 쇼핑몰 이름을 가져옵니다.
# `        product_lists.append([i+1, query, title, link, thumbnail, lprice, mall_name])`: 추출한 상품 정보를 `product_lists`에 추가합니다.
# `        print(([i+1, query, title, link, thumbnail, lprice, mall_name]))`: 각 상품 정보를 콘솔에 출력합니다.

# `    if not product_lists:`: `product_lists`가 비어 있다면 (검색된 상품이 없다면).
# `        print("[정보] 검색된 상품이 없어 DataFrame을 생성하지 않습니다.")`: 정보 메시지를 출력합니다.
# `        return pd.DataFrame(columns=['no', 'query', 'title', 'link', 'thumbnail', 'lprice', 'mall_name'])`: 빈 DataFrame을 반환합니다.

# `    df = pd.DataFrame(product_lists, columns=['no', 'query', 'title', 'link', 'thumbnail', 'lprice', 'mall_name'])`: `product_lists`를 Pandas DataFrame으로 변환합니다.
# `    df.to_csv(f'navershopping_product_lists.csv', index=False)`: DataFrame을 `navershopping_product_lists.csv` 파일로 저장합니다. 인덱스는 저장하지 않습니다.
# `    df.to_csv(f'backup/navershopping_product_lists_{timestamp}.csv', index=False)`: DataFrame을 백업용으로 타임스탬프가 포함된 파일명으로 저장합니다.

# `    return df`: 생성된 DataFrame을 반환합니다.
```

### 4. 스크립트 실행 진입점

```python
# `if __name__ == "__main__":`: 스크립트가 직접 실행될 때만 `main()` 함수를 호출하도록 합니다. (다른 모듈에서 임포트될 때는 실행되지 않음)
# `    main()`: `main` 함수를 호출하여 스크립트를 시작합니다.
```

```python
from requests import get
import json
import pandas as pd
from datetime import datetime


def get_naverapi(query):
    product_lists = []

    timestamp = datetime.now().strftime('%Y%m%d%H%M%S')
    client_id = "J2wW6Xu6fxJiS0zTlCD0"
    client_secret = "SYJBdwzack"

    user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64) "\
        "AppleWebKit/537.36 (KHTML, like Gecko) " \
        "Chrome/51.0.2704.103 Safari/537.36"
    headers = {"User-Agent": user_agent,
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret}

    url = f"https://openapi.naver.com/v1/search/shop?query={query}&display=100"
    response = get(url, headers=headers)

    try:
        response = get(url, headers=headers)
        elements = json.loads(response.text)['items']
    except Exception as e:
        print(f"[오류] {type(e).__name__}: {e}")

    for i, elm in enumerate(elements):
        title = elm['title'].replace('<b>','').replace('</b>','')
        link = elm['link']
        thumbnail = elm['image']
        lprice = elm['lprice']
        mall_name = elm['mallName']
        product_lists.append([i+1, query, title, link, thumbnail, lprice, mall_name])
        print(([i+1, query, title, link, thumbnail, lprice, mall_name]))

    df = pd.DataFrame(product_lists, columns=['no', 'query', 'title', 'link', 'thumbnail', 'lprice', 'mall_name'])
    df.to_csv(f'navershopping_product_lists.csv', index=False)
    df.to_csv(f'backup/navershopping_product_lists_{timestamp}.csv', index=False)

    return df


def main():
    try:
        query = "케이팝굿즈"
        get_naverapi(query)
    except Exception as e:
        print(f"[오류] {type(e).__name__}: {e}")

    return


if __name__ == "__main__":
    main()
```
